层次聚类（Hierarchical Clustering）

1. 导入库与数据

```python
## 导入库
import numpy as np
import matplotlib.pyplot as plt
import scipy.cluster.hierarchy as sch
from sklearn.cluster import AgglomerativeClustering
from sklearn.datasets import make_blobs

## 生成数据
X, _ = make_blobs(n_samples=10, centers=3, random_state=42)  ## 生成10个点，3个簇
X  ## 因为是非监督学习，只关注特征，不关注label
```

2. 绘制树状图

```python
plt.figure(figsize=(8, 5))
## sch.linkage用于创建一个层次结构的链接矩阵，ward方法最小化簇内的方差
## sch.dendrogram可视化聚类过程
dendrogram = sch.dendrogram(sch.linkage(X, method='ward'))
plt.title("Dendrogram for Hierarchical Clustering")
plt.xlabel("Data Points")
plt.ylabel("Distance")
plt.show()
```

3. 进行层次聚类

```python
## 使用 Ward 最小方差法进行聚类，指定需要3个簇
hc = AgglomerativeClustering(n_clusters=3, linkage='ward')
labels = hc.fit_predict(X) ## 返回聚类标签(0,1,2)
```

4. 可视化聚类结果

```python
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis', edgecolors='k')
plt.title("Hierarchical Clustering Result")
plt.show()
```
