1. 数据预处理

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

dataset = pd.read_csv('./data/studentscores.csv')
X = dataset.iloc[ : , : 1 ].values  ## 选取第0列作为x_feature
Y = dataset.iloc[ : , 1 ].values  ## 选取第1列作为y

from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split( X, Y, test_size = 1/4, random_state = 0) 
```

2. 训练集使用简单线性回归模型来训练

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression() ## 创建一个LinearRegression实例
regressor = regressor.fit(X_train, Y_train) ## 对训练集上的X和Y进行线性拟合
```

3. 预测结果

```python
Y_pred = regressor.predict(X_test)  ## 通过训练集得到的关系，预测在测试集上的结果
```

4. 可视化

```python
## 训练集结果可视化
plt.scatter(X_train , Y_train, color = 'red')
plt.plot(X_train , regressor.predict(X_train), color ='blue') ## 绘制拟合曲线
plt.show()
```

```python
## 测试集结果可视化
plt.scatter(X_test , Y_test, color = 'red')
plt.plot(X_test , regressor.predict(X_test), color ='blue')
plt.show()
```
